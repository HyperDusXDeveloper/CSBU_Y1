6. จงเขียนโปรแกรมคำนวณหาพื้นที่สามเหลี่ยม  [Area = 0.5 * base * height]  โดยมีการรับและแสดงผลดังตัวอย่าง

```
Enter base : 3.5
Enter height : 8
Area =   14.0
```


````python
base = float(input("Enter base : "))
height = float(input("Enter height : "))
Area = 0.5 * base * height
print("Area = 0.5 * ",base,"*",height,"=",Area)
````

7. การแปลงค่าระยะทางจากกิโลเมตร เป็นไมล์ [Miles = Kilometers * 0.621371] โดยมีการรับและแสดงผลดังตัวอย่าง

```
Enter kilometers : 6
Convert  6 kilometers =  3.7282260000000003 miles
```


````python
kilometers = int(input("Enter kilometers : "))
miles = kilometers * 0.621371
print("Convert",kilometers,"kilometers =","%.3f"%miles,"miles")
````

8. จงเขียนโปรแกรมทำการแปลงอุณหภูมิจากองศาเซลเซียส เป็นฟาเรนไฮท์  [Fahrenheit = (Celsius * 1.8) + 32 ] โดยมีการรับและแสดงผลดังตัวอย่าง  

```
Enter celsius : 38.5
38.5 celsius =  101.3  fahrenheit
```

````python
celsius = float(input("Enter celsius : "))
fahrenheit =(celsius * 1.8)+32
print(celsius,"celsius = ",fahrenheit,"fahrenheit")
````

9. จงเขียนโปรแกรมหาค่าดัชนีมวลกาย (BMI) โดยใช้สูตรทั่วไปที่ใช้ได้กับทุกเพศ
    [BMI = (weight / height2) ] โดยมีการรับและแสดงผลดังตัวอย่าง

```
Enter weight (kg.): 48
Enter height (m.): 1.52
BMI = 20.78
```

````python
weight = int(input("Enter weight (Kg.) : "))
height = float(input("Enter height (m.): "))
BMI = (weight / height**2)
print("BMI =%.2f " %BMI)
````
